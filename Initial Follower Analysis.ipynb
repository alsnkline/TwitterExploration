# Exploring 6000 followers of @GOP and @TheDemocrats
Initial look to see if there are going to be 500 active users - need 5 other indicative accounts also

```python
import pandas as pd
import os

import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
from IPython.display import Markdown, display, HTML

cwd = os.getcwd()
print(cwd)
```

I've previously collecting data sets of 6000 followers for @GOP and @TheDemocrats. Lets load that data:

```python
rep = pd.read_csv(cwd+'/data/followersOfGOP.csv')
dem = pd.read_csv(cwd+'/data/followersOfTheDemocrats.csv')
```

Exploring the columns of both data sets:

```python
print('Total columns in rep date = {} and in dem data = {}'.format(rep.shape[1], dem.shape[1]))
in_both = set(rep.columns.tolist()).intersection(dem.columns.tolist())
print('These {} columns are in both files:'.format(len(in_both)))
print('    {}'.format('\n    '.join(in_both)))
```

statuses_count are the number of tweets a user has ever put out and favourites_count is the number of tweets the user has favorited. Exploring distributions of these:

```python
plt.rcParams["figure.figsize"] = (20,8)

plt.subplot(2,2,1)
plt.hist(rep['statuses_count'], bins=100, log=True)
plt.title('Republican Data, Status Counts')

plt.subplot(2,2,2)
plt.hist(dem['statuses_count'], bins=100, log=True)
plt.title('Democrat Data, Status Counts')

plt.subplot(2,2,3)
plt.hist(rep['favourites_count'], bins=100, log=True)
plt.title('Republican Data, Favorite Counts')

plt.subplot(2,2,4)
plt.hist(dem['favourites_count'], bins=100, log=True)
plt.title('Democrat Data, Favorite Counts')

plt.show()
```

Republicans look to be a lot more active on Twitter!

Lets look at more information from some of the more active accounts:

```python
active_threshold = 1000
isGOPActive = rep['statuses_count'] > active_threshold
isDemActive = dem['statuses_count'] > active_threshold
inc_columns = ['name', 'screen_name','description','favourites_count','statuses_count']

display(Markdown('**first 20 of the {} qualifing active followers of @GOP:**'.format(len(rep.loc[isGOPActive,inc_columns]))))
display(rep.loc[isGOPActive,inc_columns].head(20))
display(Markdown('**first 20 of the {} qualifing active followers of @TheDemocrats:**'.format(len(dem.loc[isDemActive,inc_columns]))))
display(dem.loc[isDemActive,inc_columns].head(20))
```
